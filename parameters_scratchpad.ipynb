````python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
````

````python
kidpan_data = pd.read_csv(
    "C:/Users/brand/Desktop/STAR_Delimited/Delimited Text File 202312/Kidney_ Pancreas_ Kidney-Pancreas/KIDPAN_DATA.DAT",
    sep="\t"
)
````

````python
colnames = pd.read_csv(
    "C:/Users/brand/Desktop/STAR_Delimited/Analysis/KIDPAN_DATA_colnames.csv",
    sep=",",
    header=0
)
````

````python
kidpan_data.columns = list(colnames["LABEL"])
````

````python
kidpan_data = kidpan_data[kidpan_data["WL_ORG"] == "KI"]
kidpan_data.shape
````

````python
ages = kidpan_data["INIT_AGE"].replace(".", np.nan).astype(float)
ax = ages.hist(bins=np.arange(0,np.max(ages)+10,10))
ax.set_xlabel("Age")
````

````python
init_cpra = kidpan_data["INIT_CPRA"].replace(".", np.nan).astype(float)
ax = ages.hist(bins=np.arange(0,np.max(init_cpra)+10,10))
ax.set_xlabel("Initial CPRA (among patients)")
````

````python
end_cpra = kidpan_data["END_CPRA"].replace(".", np.nan).astype(float)
(init_cpra == end_cpra)[~np.isnan(init_cpra) & ~np.isnan(end_cpra)].value_counts()
````
